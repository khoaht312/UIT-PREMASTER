**Big Five Model** là mô hình tâm lý được nghiên cứu và phát triển bởi nhiều nhà khoa học trên thế giới. Mô hình này cho rằng trong tính cách của một người đều bao gồm 5 yếu tố và nhiệm vụ của trắc nghiệm tâm lý là xác định mức độ biểu hiện của từng yếu tố này. Big Five Model được đánh giá cao về độ tin cậy và tính khoa học. Mô hình đặc điểm tính cách Big Five đo lường 5 nét tính cách của con người: <br>
- Tự chủ tận tâm (Coscientiouness)
- Hướng ngoại (Extraversion)
- Hòa đồng (Agreeableness)
- Sẵn sàng trải nghiệm (Openness to Experience)
- Bất ổn cảm xúc (Neuroticism). <br>

Nên còn được gọi là mô hình **OCEAN** <br>

|Yếu tố| Chỉ số cao | Chỉ số thấp|
|-|-|-|
|Sẵn sàng trải nghiệm (O)|Người có điểm cao ở yếu tố này thường là người thích những ý tưởng mới, thích hiểu biết nhiều lĩnh vực nhưng đồng thời cũng thích tự do, không thích bị ràng buộc…|Người có điểm thấp ở yếu tố này thường khá bảo thủ, khó tiếp nhận những ý tưởng mới, lạ. Họ thích sự ổn định, quen thuộc và thực tế.|
|Tự chủ tận tâm (C)|Người có điểm cao ở yếu tố này thường là người chăm chỉ, có khả năng chịu áp lực tốt. Họ thường là người gắn bó, trung thành với tổ chức.|Người có điểm thấp ở yếu tố này thường dễ bỏ cuộc, khả năng chịu áp lực, tuân thủ kỷ luật của tổ chức kém|
|Hướng ngoại (E)|Người có điểm cao ở yếu tố này thường là người nhiệt tình, năng động, giao tiếp tốt, thích thể hiện bản thân.|Người có điểm thấp ở yếu tố này thường ngại giao tiếp, không thích sự nổi bật, thích được làm việc độc lập.|
|Hòa đồng dễ chịu (A)|Người có điểm cao ở yếu tố này thường thân thiện, cởi mở, đồng cảm với mọi người nhưng nhiều khi “thiếu chính kiến"|Người có điểm thấp thường đặt lợi ích của bản thân lên trên, ít đồng cảm, chia sẻ với đồng nghiệp, có tính cạnh tranh cao|
|Bất ổn cảm xúc (N)|Người có điểm cao ở yếu tố này thường có các cảm xúc tiêu cực như: lo lắng, bực bội, tự ti, yếu đuối và khả năng chịu áp lực kém|Người có điểm thấp ở yếu tố này thường kiểm soát được cảm xúc, ứng phó với căng thẳng tốt, ít bị bên ngoài ảnh hưởng đến tâm trạng của bản thân.|

C74 có nghĩa là 74% của những người đã thử nghiệm đạt điểm thấp hơn so với bạn trong sự tận tâm. Vì vậy, bạn đang khá có tổ chức và kỷ luật tự giác, so với những người khác đã thử nghiệm.

**Các tính chất của OOP**:

**Tính kế thừa (Inheritance)**: Các lớp con O, C, E, A, N được kế thừa từ lớp cơ sở Yeuto, thể hiện tính kế thừa trong OOP.

**Tính đa hình (Polymorphism)**: Các phương thức như Xuat() và NguyCoCao() được định nghĩa lại trong các lớp con, trường hợp lớp O,A không có định nghĩa lại NguyCoCao() nên thực hiện hàm ở lớp cha (YeuTo).

**Tính đóng gói (Encapsulation)**: Dữ liệu được đóng gói trong các lớp, ví dụ như biến iChiSo trong lớp Yeuto, giúp ẩn thông tin và chỉ truy cập thông qua các phương thức.

**Sử dụng cấu trúc lớp (Class Structure)**: Code sử dụng cấu trúc lớp để tổ chức và mô tả các đặc tính và hành vi của các đối tượng.

**Tính trừu tượng (Abstraction)**: Các lớp cung cấp một mức độ trừu tượng thông qua việc mô hình hóa các đặc điểm chung của các loại người dựa trên tính chất.

```python
class Yeuto:
    def __init__(self):
        self.iChiSo = 0

    def NguyCoCao(self):
        return 0

    def Nhap(self):
        self.iChiSo = int(input("Nhập chỉ số: "))

    def Xuat(self):
        print("\n", self.iChiSo, " ", end="")


class O(Yeuto):
    def __init__(self):
        super().__init__()

    def Xuat(self):
        super().Xuat()
        print("| Sẵn sàng trải nghiệm (O). ", end="")
        if self.iChiSo >= 70:
            print("Người có điểm cao ở yếu tố này thường là người thích những ý tưởng mới, thích hiểu biết nhiều lĩnh vực nhưng đồng thời cũng thích tự do, không thích bị ràng buộc…")
        elif self.iChiSo > 30 and self.iChiSo < 70:
            print("Không xác định rõ")
        else:
            print("Người có điểm thấp ở yếu tố này thường khá bảo thủ, khó tiếp nhận những ý tưởng mới, lạ. Họ thích sự ổn định, quen thuộc và thực tế.")


class C(Yeuto):
    def __init__(self):
        super().__init__()

    def Xuat(self):
        super().Xuat()
        print("| Tự chủ tận tâm (C). ", end="")
        if self.iChiSo >= 70:
            print("Người có điểm cao ở yếu tố này thường là người chăm chỉ, có khả năng chịu áp lực tốt. Họ thường là người gắn bó, trung thành với tổ chức.")
        elif self.iChiSo > 30 and self.iChiSo < 70:
            print("Không xác định rõ")
        else:
            print("Người có điểm thấp ở yếu tố này thường dễ bỏ cuộc, khả năng chịu áp lực, tuân thủ kỷ luật của tổ chức kém.")

    def NguyCoCao(self):
        if self.iChiSo <= 30:
            return 1
        return 0


class E(Yeuto):
    def __init__(self):
        super().__init__()

    def Xuat(self):
        super().Xuat()
        print("| Hướng ngoại (E). ", end="")
        if self.iChiSo >= 70:
            print("Người có điểm cao ở yếu tố này thường là người nhiệt tình, năng động, giao tiếp tốt, thích thể hiện bản thân.")
        elif self.iChiSo > 30 and self.iChiSo < 70:
            print("Không xác định rõ")
        else:
            print("Người có điểm thấp ở yếu tố này thường ngại giao tiếp, không thích sự nổi bật, thích được làm việc độc lập.")

    def NguyCoCao(self):
        if self.iChiSo <= 30:
            return 1
        return 0


class A(Yeuto):
    def __init__(self):
        super().__init__()

    def Xuat(self):
        super().Xuat()
        print("| Hòa đồng dễ chịu (A). ", end="")
        if self.iChiSo >= 70:
            print("Người có điểm cao ở yếu tố này thường thân thiện, cởi mở, đồng cảm với mọi người nhưng nhiều khi thiếu chính kiến")
        elif self.iChiSo > 30 and self.iChiSo < 70:
            print("Không xác định rõ")
        else:
            print("Người có điểm thấp thường đặt lợi ích của bản thân lên trên, ít đồng cảm, chia sẻ với đồng nghiệp, có tính cạnh tranh cao.")

    def NguyCoCao(self):
        if self.iChiSo >= 70:
            return 1
        return 0


class N(Yeuto):
    def __init__(self):
        super().__init__()

    def Xuat(self):
        super().Xuat()
        print("| Bất ổn cảm xúc (N). ", end="")
        if self.iChiSo >= 70:
            print("Người có điểm cao ở yếu tố này thường có các cảm xúc tiêu cực như: lo lắng, bực bội, tự ti, yếu đuối và khả năng chịu áp lực kém.")
        elif 30 < self.iChiSo < 70:
            print("Không xác định rõ")
        else:
            print("Người có điểm thấp ở yếu tố này thường kiểm soát được cảm xúc, ứng phó với căng thẳng tốt, ít bị bên ngoài ảnh hưởng đến tâm trạng của bản thân.")


class Nguoi:
    def __init__(self):
        self.arrOcean = [O(), C(), E(), A(), N()]

    def Nhap(self):
        for i in range(5):
            self.arrOcean[i].Nhap()

    def Xuat(self):
        for i in range(5):
            self.arrOcean[i].Xuat()

    def NguyCoCao(self):
        iC = self.arrOcean[1].NguyCoCao()
        iE = self.arrOcean[2].NguyCoCao()
        iN = self.arrOcean[4].NguyCoCao()
        if iC == 1 or iN == 1 or (iE == 1 and iN == 1):
            return 1
        return 0


class MangNguoi:
    def __init__(self):
        self.arrNguoi = [Nguoi() for _ in range(100)]
        self.iN = 0

    def Nhap(self):
        self.iN = int(input("Nhập số lượng người: "))
        for i in range(self.iN):
            print("Nhập các chỉ số (O-C-E-A-N) của người thứ ", i + 1, ":")
            self.arrNguoi[i].Nhap()

    def Xuat(self):
        for i in range(self.iN):
            print("Kết quả đánh giá tâm lý Big Five của người thứ ",i + 1, " như sau:")
            self.arrNguoi[i].Xuat()

    def NguyCoCao(self):
        print("Cảnh báo:")
        for i in range(self.iN):
            if self.arrNguoi[i].NguyCoCao() == 1:
                print("Người thứ {} là người có nguy cơ cao mà nhà tuyển dụng/ tổ chức/ doanh nghiệp cần lưu ý.".format(i+1))


def main():
    a = MangNguoi()
    a.Nhap()
    a.Xuat()
    a.NguyCoCao()


if __name__ == "__main__":
    main()
```
